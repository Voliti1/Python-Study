## Variable(변수)

값을 저장하기 위한 메모리 공간

## Data Type

**Int**
- 정수 저장하는 자료형
- ex) 0, -1, 1

**Float**
- 실수 저장하는 자료형
- Float는 Floating Point(부동소수점)에서 따옴
- ex) 0.0, -1.2, 123.456
