# Python


## Variable(변수)

값을 저장하기 위한 메모리 공간

## Data Type(자료형)
### Number

**Int**
- 정수 저장하는 자료형
- ex) 0, -1, 1

**Float**
- 실수 저장하는 자료형
- Float는 Floating Point(부동소수점)에서 따옴
- ex) 0.0, -1.2, 123.456
- 지수표현 방식도 사용 가능

# Operator(연산자)

\+ : 더하기<br>
\- : 빼기<br>
\* : 곱하기<br>
/ : 나누기<br>
// : 몫 반환<br>
% : 나머지 반환<br>
** : 제곱<br>

```python
a = 3; b = 2
print("a =", a ,"b =", b)
print("a + b =", (a + b))
print("a - b =", (a - b))
print("a * b =", (a * b))
print("a / b =", (a / b))
print("a // b =", (a // b))
print("a % b =", (a % b))
print("a ** b =", (a ** b))
```
