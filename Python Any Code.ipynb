```python
score = int(input("정수를 입력하세요 : "))

if score >= 60:
    print(str(score) + "점 합격입니다.")
else:
    print(str(score) + "점 불합격입니다.")
```

```python
math = int(input('수학 점수를 입력해 주세요 : '))
eng = int(input('수학 점수를 입력해 주세요 : '))

if math >= 60 and eng >= 60:
    print(str(math) + "점", str(eng) + "점, 합격입니다.")
else:
    print(str(math) + "점", str(eng) + "점, 불합격입니다.")
```

```python
pocket = ['phone', 'wallet', 'money', 'IDcard']

if 'money' in pocket:
    print("택시를 타세요")
else:
    print("걸어가세요")

if 'money' not in pocket:
    print("걸어가세요")
else:
    print("택시를 타세요")
```

```python
a = int(input("첫 번째 정수값을 입력해 주세요 : "))
b = int(input("두 번째 정수값을 입력해 주세요 : "))

if (a > b):
    print("첫 번째 입력한 숫자가 더 큽니다. 첫 번째 숫자 : %d" %a)
elif (a < b):
    print("두 번째 입력한 숫자가 더 큽니다. 두 번째 숫자 : %d" %b)
else:
    print("같습니다.")
```
